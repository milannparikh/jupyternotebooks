# Welcome to my <i>Caesar Cipher</i>!!!

This first section opens a wordlist file and breaks the words into a list.

```python
file = open("wordlist.txt",'r')
wordlist = file.read()
words = (list(wordlist.split()))
file.close()
```

Now we create a table for the alphabet.

```python
table = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
```

Now a while loop for the interface:

```python
boole = True
while boole == True:
    # Provides basic options and breaks the code into two major categories: encoding and decoding.
    # Or is it encrypting and decrypting? I really don't know.
    temp = input('Do you want to encrypt (e), decrypt (d), or stop (q)? ')

    # Path for encryption 
    if temp == 'e':
        rtable = []
        string = input('What message would you like to encrypt? ')
        key = int(input('How far would you like to rotate the cipher? Enter a number between 1 and 25: '))

        # Creates a new table (rtable) starting with index 0 and adding the key.
        # It then continues through the alphabet.
        # The % (modulo) allows 'a' to come after 'z'.
        for i in range(len(table)):
            rtable.append(table[(i+key)%26])

        # This starts building the new message.
        # It finds the index of each letter in the old table, then pulls the letter in that index in the new table.
        answer = ''
        for i in range(len(string)):

            # Provides exceptions for spaces and punctuation.
            if string[i].lower() not in table:
                answer = answer + str(string[i])

            # Allows capital letters to stay capital.
            elif ((string[i] not in table) and (string[i].lower() in table)):
                answer = answer + str(rtable[table.index(string[i].lower())].upper())

            # Takes care of the lowercase letters.
            else:
                answer = answer + str(rtable[table.index(string[i])])

        print('Your encrypted message is: ' + answer)

    # Path for decryption
    elif (temp == 'd'):
        # Provides three options for decryption
        option = input('Would you like to use a key (k), a piece of decrypted text (t), or nothing at all (n)? ')

        # The simplest option: using the key to find the original message.
        if option == 'k':
            cipher = input('What message would you like to decrypt? ')
            key = int(input('What is the key? '))

            # I use the same code as the encryption segment to decrypt things.
            # I have to make sure it rotates the proper amount from where it is currently.
            rot = abs(26-key)

            rtable = []
            for i in range(26):
                rtable.append(table[(i+rot)%26])
            original = ''
            for i in range(len(cipher)):
                if cipher[i].lower() not in table:
                    original = original + str(cipher[i])
                elif ((cipher[i] not in table) and (cipher[i].lower() in table)):
                    original = original + str(rtable[table.index(cipher[i].lower())].upper())
                else:
                    original = original + str(rtable[table.index(cipher[i])])
            print('Here is the original message: ' + original)

        # This option allows the user to input a piece of decrypted text from the message.
        elif option == 't':
            cipher = input('What message would you like to decrypt? ')
            text = input('Can I have some decrypted text from the cipher to work with? ')

            # This sets artificial keys for the program to run through.
            for p in range(1, 26):
                rtable = []
                for i in range(26):
                    rtable.append(table[(i+p)%26])
                original = ''
                for i in range(len(cipher)):
                    if cipher[i].lower() not in table:
                        original = original + str(cipher[i])
                    elif ((cipher[i] not in table) and (cipher[i].lower() in table)):
                        original = original + str(rtable[table.index(cipher[i].lower())].upper())
                    else:
                        original = original + str(rtable[table.index(cipher[i])])

                # This searches for the decrypted text in each possible message.
                # If it finds it it will print it to the user along with the key used to encrypt the message.
                # If the same text is found with multiple keys, it will print all messages it finds.
                if text in original:
                    print('Here is the original message: ' + original)
                    print('The key was ' + str(abs(26-p)) + '.')

        # This is the cool part. Using English word lists to find the most likely answer from an encrypted message.
        elif option == 'n':
            cipher = input('What message would you like to decrypt? ')
            numwords = []
            messages = []
            # This folllows the same procedure as the last option as we don't know the key.
            for p in range(1, 26):
                rtable = []
                for i in range(26):
                    rtable.append(table[(i+p)%26])
                original = ''
                for i in range(len(cipher)):
                    if cipher[i].lower() not in table:
                        original = original + str(cipher[i])
                    elif ((cipher[i] not in table) and (cipher[i].lower() in table)):
                        original = original + str(rtable[table.index(cipher[i].lower())].upper())
                    else:
                        original = original + str(rtable[table.index(cipher[i])])

                # This breaks each message into individual words.
                sentence = original.split()
                rsentence = []

                # This takes all punctuation out of each word
                for i in sentence:
                    if len(i.split()) > 1:
                        rsentence.append(''.join(c for c in i if c in table))
                    elif len(i.split()) == 1:
                        rsentence.append(i)

                # This counts the number of English words in the message.
                num = 0
                for word in rsentence:
                    if word.lower() in words:
                        num += 1

                # This adds that number to a list in order of the key used.
                numwords.append(num)

                # This appends each message formed with each key to a separate list.
                messages.append(original)

            # Finds the indices of the messages with the max number of English words.
            maximum = max(numwords)
            indices = [i for i, j in enumerate(numwords) if j == maximum]
            # ^-- What does this actually do? I know it helps find indices but I don't fully understand.
            
            for q in indices:
                print('Your message could be: ' + str(messages[q]))
                print('The key is ' + str(abs(26-q-1)) + '.\n')

        else:
            print('That isn\'t a valid option. Try again.')

    # Breaks while loop.
    elif(temp == 'q'):
        print('Goodbye!')
        break

    else:
        print('Nice try but not a valid option. Try again')
```
